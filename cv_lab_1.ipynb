#Task 1

```python


def display_menu():
    print("\nGrocery List Management System")
    print("1. Add an item")
    print("2. Remove an item")
    print("3. Display the list")
    print("4. Exit")

def add_item(grocery_list):
    item = input("Enter the item to add: ")
    grocery_list.append(item)
    print(f"'{item}' has been added to the list.")

def remove_item(grocery_list):
    item = input("Enter the item to remove: ")
    if item in grocery_list:
        grocery_list.remove(item)
        print(f"'{item}' has been removed from the list.")
    else:
        print(f"'{item}' is not in the list.")

def display_list(grocery_list):
    if not grocery_list:
        print("The grocery list is empty.")
    else:
        print("\nCurrent Grocery List:")
        for i, item in enumerate(grocery_list, start=1):
            print(f"{i}. {item}")

def main():
    grocery_list = []
    while True:
        display_menu()
        choice = input("Enter your choice (1-4): ")

        if choice == '1':
            add_item(grocery_list)
        elif choice == '2':
            remove_item(grocery_list)
        elif choice == '3':
            display_list(grocery_list)
        elif choice == '4':
            print("Exiting the program. Goodbye!")
            break
        else:
            print("Invalid choice. Please choose a number between 1 and 4.")

if __name__ == "__main__":
    main()
```

#Task 2

```python
# Student Record System

def display_menu():
    print("\nStudent Record System")
    print("1. Add a student")
    print("2. Update a student's grade")
    print("3. Display all student records")
    print("4. Exit")

def add_student(student_records):
    student_id = input("Enter the student ID: ")
    if student_id in student_records:
        print(f"Student with ID {student_id} already exists.")
    else:
        name = input("Enter the student's name: ")
        grade = input("Enter the student's grade: ")
        student_records[student_id] = {'name': name, 'grade': grade}
        print(f"Student {name} with ID {student_id} has been added.")

def update_grade(student_records):
    student_id = input("Enter the student ID to update: ")
    if student_id in student_records:
        new_grade = input(f"Enter the new grade for {student_records[student_id]['name']}: ")
        student_records[student_id]['grade'] = new_grade
        print(f"Grade updated for student ID {student_id}.")
    else:
        print(f"No student found with ID {student_id}.")

def display_students(student_records):
    if not student_records:
        print("No student records available.")
    else:
        print("\nStudent Records:")
        for student_id, details in student_records.items():
            print(f"ID: {student_id}, Name: {details['name']}, Grade: {details['grade']}")

def main():
    student_records = {}
    while True:
        display_menu()
        choice = input("Enter your choice (1-4): ")

        if choice == '1':
            add_student(student_records)
        elif choice == '2':
            update_grade(student_records)
        elif choice == '3':
            display_students(student_records)
        elif choice == '4':
            print("Exiting the program. Goodbye!")
            break
        else:
            print("Invalid choice. Please choose a number between 1 and 4.")

if __name__ == "__main__":
    main()
```

```python
image = cv2.imread('/content/sample_image.jpeg', cv2.IMREAD_GRAYSCALE)


ret, thresholded_image = cv2.threshold(image, 127, 255, cv2.THRESH_BINARY)
height, width = thresholded_image.shape
center = (width // 2, height // 2)
rotation_matrix = cv2.getRotationMatrix2D(center, 60, 1.0)
rotated_image = cv2.warpAffine(thresholded_image, rotation_matrix, (width, height))
print('original image')
cv2_imshow(image)
print('rotated image')
cv2_imshow(rotated_image)
```
