```python
from ultralytics import YOLO

# Load pre-trained model (you can also use yolov8s.pt or yolov8m.pt, etc.)
model = YOLO("/content/yolov8n.pt")

# Train the model using the extracted data
model.train(data='/content/Construction_Site_Safety_yolov8/data.yaml', epochs=29, imgsz=640)
```

```python
from ultralytics import YOLO
from PIL import Image

# Load the trained model
model = YOLO('/content/runs/detect/train/weights/best.pt')

# Predict on the image
results = model.predict('/content/constructionsafety.jpg')

# Display results
results[0].show()  # shows the image with bounding boxes
```

```python
import matplotlib.pyplot as plt
from PIL import Image

# Run prediction
results = model.predict('/content/constructionsafety.jpg')

# Get the rendered image with boxes
rendered_img = results[0].plot()  # this returns a NumPy array

# Show bigger image
plt.figure(figsize=(15, 10))  # adjust size here
plt.imshow(rendered_img)
plt.axis('off')
plt.show()
```
